A partir del Espacio ROC ya se puede analizar el rendimiento de cada clasificador respecto al dataset considerado, teniendo en cuenta su posición respecto a la diagonal, que representa una clasificación completamente aleatoria.

**Comparar Naive Bayes, Vecinos Próximos y Regresión Logística**

Para pima-indian-diabetes, no da muy buenos resultados, ya que los 3 clasificadores se alejan mucho de (0, 1). La mejor sería KNN, pero tampoco podríamos decir que es muy exacto, ya que tu tasa de falsos positivos es muy baja pero si tasa de verdaderps posritivos es muy cercana a 50%.
Naive bayes es la mas cercana a ser clasificada en un punto donde lo considera un clasificador que clasifica de forma aleatoria. La tasa de falsos positivos en RL es 100 y la tasa de verdaderos positivos es 0.

Para wdbc, la peor es regeresion logistica, donde la tasa de verdaderos positivos es 0, (al igual que KNN). La tasa de falsos positivos en RL es 100%, totalmente contraria a KNN, que da 0 (clasifica como positivo de forma errónea menos veces), al igual que su tasa de verdaderos positivos (clasifica como positivo correctamente muy pocas veces). Es tan bajo en KNN que lo clasifica en un punto donde lo considera un clasificador que clasifica de forma aleatoria. La mejor es NB, que estas cerca del punto (0, 1), que seria un clasificador perfecto.
