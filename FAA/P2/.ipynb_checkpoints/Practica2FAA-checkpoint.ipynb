Varía con cada ejecucion, pero podemos observar que como mucho hace 5 iteraciones. Todo depende de los clustyers que quieras tener. Cuanto mas alejados estan, (4 y 5), mas diferencias entre las plantas encuentra, por lo que la clasificacion desvaría mas de lo que debería ser.

Este problema se resuelve bien con K=3, ya que hay 3 clases de flores.
